1 — Write Products into CSV File

```

import csv

products = [
    [101, "Laptop", "Electronics", 75000, 10],
    [102, "Mouse", "Accessories", 800, 50],
    [103, "Keyboard", "Accessories", 1500, 30],

]

file = open("products.csv", "w", newline="")

writer = csv.writer(file)

writer.writerow(["ID", "Name", "Category", "Price", "Quantity"])

writer.writerows(products)

file.close()

print("CSV file created successfully")
```

2 — Read Data from CSV File

```
import csv

file = open("products.csv", "r")

reader = csv.reader(file)

for row in reader:
    print(row)

file.close()
```

3 — Delete Product from CSV File

```
import csv

rows = []

file = open("products.csv", "r")

reader = csv.reader(file)

for row in reader:
    if row[0] != "103":
        rows.append(row)

file.close()

file = open("products.csv", "w", newline="")

writer = csv.writer(file)

writer.writerows(rows)

file.close()

print("Product deleted")
```

4 — Display Products with Price Greater than 5000

```
import csv

file = open("products.csv", "r")

reader = csv.reader(file)

next(reader)

for row in reader:
    if int(row[3]) > 5000:
        print(row)

file.close()
```

5 — Write Products into JSON File

```
import json

products = [
    {"id": 101, "name": "Laptop", "category": "Electronics", "price": 75000, "quantity": 10},
    {"id": 102, "name": "Mouse", "category": "Accessories", "price": 800, "quantity": 50},
    {"id": 103, "name": "Keyboard", "category": "Accessories", "price": 1500, "quantity": 30},
    {"id": 104, "name": "Mobile", "category": "Electronics", "price": 25000, "quantity": 15},
    {"id": 105, "name": "Headphones", "category": "Accessories", "price": 2000, "quantity": 20}
]

file = open("products.json", "w")

json.dump(products, file, indent=4)

file.close()

print("JSON file created")
```

6 — Read Data from JSON File

```
import json

file = open("products.json", "r")

data = json.load(file)

for product in data:
    print(product)

file.close()
```

7 — Delete Product from JSON File

```
import json

file = open("products.json", "r")

data = json.load(file)

file.close()

new_data = []

for product in data:
    if product["id"] != 102:
        new_data.append(product)

file = open("products.json", "w")

json.dump(new_data, file, indent=4)

file.close()

print("Product deleted")
```

8 — Update Product Quantity

```
import json

file = open("products.json", "r")

data = json.load(file)

file.close()

for product in data:
    if product["name"] == "Laptop":
        product["quantity"] = 25

file = open("products.json", "w")

json.dump(data, file, indent=4)

file.close()

print("Quantity updated")
```

9 — Display Electronics Products Only

```
import json

file = open("products.json", "r")

data = json.load(file)

for product in data:
    if product["category"] == "Electronics":
        print(product)

file.close()
```
