```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

medical_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Medicine_Distributed': [1200, 1500, 1800, 1700, 2100, 2500],
    'Hospital_Orders': [30, 45, 50, 48, 60, 75],
    'Delivery_Time': [5, 4, 6, 5, 3, 2],
    'Revenue': [50000, 65000, 72000, 70000, 85000, 95000]
}

df = pd.DataFrame(medical_data)

print(df)
```

Line Chart Problems


1. Create a line chart showing medicine distribution over months

```
plt.plot(df['Month'], df['Medicine_Distributed'], marker='o')

plt.title("Medicine Distribution Over Months")
plt.xlabel("Month")
plt.ylabel("Medicine Distributed")
plt.grid(True)

plt.savefig("medicine_distribution.png")

plt.show()
```

2. Create a line chart for monthly revenue growth.


```
plt.plot(df['Month'], df['Revenue'], marker='o')

plt.title("Monthly Revenue Growth")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)

plt.savefig("revenue_growth.png")

plt.show()
```

3. Create multiple line charts for Medicine_Distributed and Hospital_Orders.


```
plt.plot(df['Month'], df['Medicine_Distributed'],
          label='Medicine Distributed')

plt.plot(df['Month'], df['Hospital_Orders'],
          label='Hospital Orders')

plt.title("Medicine Distribution and Hospital Orders")
plt.xlabel("Month")
plt.ylabel("Values")

plt.legend()
plt.grid(True)

plt.savefig("multiple_line_chart.png")

plt.show()
```

4. Add markers, grid, title, and axis labels to the charts.

```
plt.plot(df['Month'], df['Medicine_Distributed'],
         marker='o', label='Medicine Distributed')

plt.plot(df['Month'], df['Hospital_Orders'],
         marker='o', label='Hospital Orders')

plt.title("Medicine Distribution and Hospital Orders")
plt.xlabel("Month")
plt.ylabel("Values")

plt.legend()
plt.grid(True)

plt.savefig("multiple_line_chart.png")

plt.show()
```

5. Create a vertical bar chart for hospital orders by month.


```
plt.bar(df['Month'], df['Hospital_Orders'])

plt.title("Hospital Orders by Month")
plt.xlabel("Month")
plt.ylabel("Orders")

plt.savefig("vertical_bar_chart.png")

plt.show()
```

6.Create a horizontal bar chart for revenue by month.


```
plt.barh(df['Month'], df['Revenue'])

plt.title("Revenue by Month")
plt.xlabel("Revenue")
plt.ylabel("Month")

plt.savefig("horizontal_bar_chart.png")

plt.show()
```

7.Create a grouped bar chart comparing Medicine_Distributed and Hospital_Orders.


```
x = np.arange(len(df['Month']))
width = 0.35

plt.bar(x - width/2,
        df['Medicine_Distributed'],
        width,
        label='Medicine')

plt.bar(x + width/2,
        df['Hospital_Orders'],
        width,
        label='Orders')

plt.xticks(x, df['Month'])

plt.title("Medicine Distributed vs Hospital Orders")
plt.xlabel("Month")
plt.ylabel("Values")

plt.legend()

plt.savefig("grouped_bar_chart.png")

plt.show()
```
