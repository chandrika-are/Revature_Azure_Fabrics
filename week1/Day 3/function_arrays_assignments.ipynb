```
numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

# 1. Array Sum -Write a function that accepts a list of numbers and returns the sum of all elements. numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]
def array_sum(nums):
    total = 0
    for i in nums:
        total += i
    return total

print("Sum =", array_sum(numbers))

```

```
# 2. Maximum Element
def maximum_element(nums):
    largest = nums[0]
    for i in nums:
        if i > largest:
            largest = i
    return largest

print("Maximum Element =", maximum_element(numbers))
```

```
# 3. Even Numbers Counter
def even_counter(nums):
    count = 0
    for i in nums:
        if i % 2 == 0:
            count += 1
    return count

print("Even Numbers Count =", even_counter(numbers))

```

```
# 4. Reverse Array
def reverse_array(nums):
    return nums[::-1]

print("4. Reversed Array =", reverse_array(numbers))
```

```
# 5. Search Element
def search_element(nums, key):
    for i in range(len(nums)):
        if nums[i] == key:
            return i
    return -1

print("Search Element Index =", search_element(numbers, 56))
```

```
# 6. Palindrome List
def palindrome_list(nums):
    if nums == nums[::-1]:
        return True
    else:
        return False

print("Is Palindrome =", palindrome_list(numbers))
```

```
# 7. Second Largest Number
def second_largest(nums):
    first = second = -1

    for i in nums:
        if i > first:
            second = first
            first = i
        elif i > second and i != first:
            second = i

    return second

print("Second Largest Number =", second_largest(numbers))
```

```
# 8. Duplicate Elements
def duplicate_elements(nums):
    duplicates = []

    for i in nums:
        if nums.count(i) > 1 and i not in duplicates:
            duplicates.append(i)

    return duplicates

print(" Duplicate Elements =", duplicate_elements(numbers))

```

```
# 9. Function for Student Marks
def student_marks(marks):
    average = sum(marks) / len(marks)
    highest = max(marks)
    lowest = min(marks)

    return average, highest, lowest

avg, high, low = student_marks(numbers)

print("Average =", avg)
print("Highest =", high)
print("Lowest =", low)
```
