```python
#partD-12 Simulate payment retry logic where the system retries payment a maximum of 3 times if payment fails.
attempt = 1

while attempt <= 3:

    payment_status = input("Enter payment status (success/fail): ")

    if payment_status == "success":
        print("Payment Successful")
        break
    else:
        print("Payment Failed")

    attempt += 1

if attempt > 3:
    print("Maximum Retry Limit Reached")
```

```python
#partD-13 Write a program that keeps checking order status until status becomes 'Delivered' or maximum 5 attempts are reached.
status = input()
attempt = 1

while status != "Delivered" and attempt <= 5:
    print("Checking order status... Attempt", attempt)
    if attempt == 4:
        status = "Delivered"

    attempt = attempt + 1

print("Final Status:", status)
```

```python
#partD-14 Inventory restock check: Keep checking stock level every iteration and stop when stock becomes greater than 100.
stock = int(input("Enter Stock Level: "))

while stock <= 100:

    print("Current Stock:", stock)
    print("Restocking Needed")

    stock += 20

print("Stock is Greater Than 100")
print("Restocking Stopped")
```

```python
#partD-15 You are given a list of orders with order_id and amount.
orders = [
    {"order_id": 101, "amount": 120000, "status": "Completed"},
    {"order_id": 102, "amount": 45000, "status": "Cancelled"},
    {"order_id": 103, "amount": 75000, "status": "Completed"},
    {"order_id": 104, "amount": 150000, "status": "Completed"}
]

premium_count = 0
Regular_Order = 0

for order in orders:
    if order["status"] == "Cancelled":
        continue
    if order["amount"] > 100000:
        print(order["order_id"], "- Premium Order")
        premium_count += 1
    else:
        print(order["order_id"], "- Regular Order")
        Regular_Order += 1

print("Total Premium Order:", premium_count)
print("Total Regular Order:", Regular_Order)
```
