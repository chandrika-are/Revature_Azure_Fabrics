```python
import pandas as pd

df = pd.read_csv(r"C:\Users\LENOVO\OneDrive\Documents\Desktop\assignment\customers.csv")

df.head()
```

```python
df.fillna({
    'City':'Unknown',
    'Email':'Not Available',
    'Age':df['Age'].mean(),
    'TotalPurchase':0
}, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.shape
```

```python

df['JoinDate'] = pd.to_datetime(df['JoinDate'])
```

```python
total_purchase = df['TotalPurchase'].sum()

print("Total Purchase:", total_purchase)
```

```python

df.to_sql(
    name="customer_load",
    con=engine,
    if_exists="replace",
    index=False
)



print("ETL Load Completed")
```

```python

pd.read_sql("SELECT * FROM customer_load", con=engine)
```

```python
#4.Find top customers by purchase

pd.read_sql("SELECT CustomerID,CustomerName,TotalPurchase FROM Customers ORDER BY TotalPurchase DESC LIMIT 5;", con=engine)
```

```python
#5.Analyze city-wise revenue
pd.read_sql("SELECT City,SUM(TotalPurchase) AS Revenue FROM Customers GROUP BY City ORDER BY Revenue DESC;", con=engine)
```

```python
#6.Rank customers using window functions

pd.read_sql("SELECT CustomerID,CustomerName,TotalPurchase,RANK() OVER (ORDER BY TotalPurchase DESC) AS CustomerRank FROM Customers;", con=engine)
```

```python
#7. Total Customers
pd.read_sql("""
SELECT MONTH(JoinDate) AS Month,
       COUNT(*) AS NewCustomers
FROM Customers
GROUP BY MONTH(JoinDate)
ORDER BY Month
""", engine)
```

```python
#8. City-wise Customer Count
pd.read_sql("SELECT City,COUNT(*) AS CustomerCount FROM Customers GROUP BY City;", con=engine)
```

```python
#9. Fetch Above Average Customers Information
pd.read_sql("SELECT * FROM Customers WHERE TotalPurchase > ( SELECT AVG(TotalPurchase) FROM Customers );", con=engine)
```

```python
#10. find Customer Acquisition Trend  - month wise
pd.read_sql("SELECT DATE_FORMAT(JoinDate,'%Y-%m') AS Month,COUNT(*) AS NewCustomers FROM Customers GROUP BY Month ORDER BY Month;", con=engine)
```

```python
#year wise
pd.read_sql("SELECT YEAR(JoinDate) AS Year, COUNT(*) AS NewCustomers FROM Customers GROUP BY Year ORDER BY Year;", con=engine)
```
